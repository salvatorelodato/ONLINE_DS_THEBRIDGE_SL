# *Proyecto EDA*
### **Análisis exploratorio de patrones de ocupación del edificio para optimización de instalaciones**

### **Contexto del Proyecto**

Contexto del Proyecto
Ante la creciente adopción de trabajo remoto, después de la pandemia, un departamento en concreto han manifestado preocupación por la baja ocupación observada en un edificio corporativo, generando costos innecesarios relacionados con energía, climatización, seguridad y limpieza. 
En conversaciones recientes, con distintos departamentos, el departamento de Servicios señaló específicamente la necesidad de entender mejor estos patrones, ya que en aspecto general, es evidente una reducción de personal, pero no se cuenta con números en concreto, para tomar decisiones fundamentadas sobre optimización de recursos, mas que un par de archivos xml y correos de accesos diarios. Para ello, acordamos colaborar directamente con otros departamentos, encargada de administrar el sistema de control de accesos del edificio, y el control de las BDD, desde donde extraeremos los datos necesarios para nuestro análisis. Este análisis exploratorio (EDA) nos permitirá identificar claramente los patrones de ocupación, generando decisiones con dichos datos.

### **Objetivo del Proyecto**
Objetivo del Proyecto
Identificar los días y horarios con menor ocupación del edificio mediante datos de acceso, con el fin de proponer mejoras en la gestión energética y operativa, optimizando los costos relacionados con la climatización, la iluminación y la seguridad.

### **Hipótesis**
Existen días específicos (viernes, lunes, fines de semana, días antes de días no laborables o periodos vacacionales) con ocupación mínima, en los cuales mantener la operación regular completa del edificio no se considera rentable. Mediante la identificación de estos patrones de ocupación es posible calcular una cantidad mínima de recursos necesario para justificar la operación del edificio. Aplicando ajustes operativos en días de baja ocupación (reducción parcial o cierre total), se lograría un ahorro significativo en costos operativos.

### **Conclusiones y Recomendaciones**:

Identificación de patrones claros y recomendaciones para optimización operativa y energética.
Estimación económica del impacto de las propuestas.

Este EDA, se utilizara para apoyar al departamento implicado, ya que en este momento solo tenemos acceso a estos datos de forma reducida, ya que se contienen muchos datos personal y podríamos violar alguna ley de protección de datos, por lo que no utilizaremos la totalidad de la información. El objetivo es Identificar patrones claros y recomendaciones para optimizar los servicios operativos y energéticos del edificio.

### **Proximos Pasos**:
Los proximos pasos se definiran de forma en que podamos extraer informacion relevante y permitir al departamento implicado tomar decisiones en base a datos, en un futuro lo ideal seria complementar estos datos con otras tablas de la BDD. Como Eventos, Ocupacion del Parking, Ocupacion por Plantas, Gasto energetico, Relacion de Horario de Entrada y Salida etc.

## Análisis Exploratorio

### Análisis Univariante

Distribución diaria de accesos: Identificación de los días con menor ocupación semanal y mensual.

Definiremos el lunes como el primer dia de la semana y el Domingo como el 7mo. 

Métricas clave: promedio  , mediana, desviación estándar de accesos diarios.

### Análisis Bivariante

Día de la semana vs. cantidad de accesos: Correlación entre tipo de día (laboral o no) y nivel de ocupación.

Mes vs. cantidad de accesos: Análisis de estacionalidad o variaciones mensuales.

### Análisis Multivariante

Día-mes-ocupacion: Combinación de variables para entender patrones más complejos. Mapa de Calor.

Análisis predictivo simple: Estimación de días con baja ocupación futura en base a tendencias.

### Visualizaciones Relevantes

Heatmaps : Identificación visual de los días más y menos concurridos.

Histogramas: Distribución de accesos diarios.

Gráficos de líneas: Tendencias de ocupación a lo largo del tiempo.

Gráficos de barras: Comparación de accesos por dia semana.

### Conclusiones

Se identifican ciertos días de la semana laborable (como viernes o lunes) con una reducción constante en el número de accesos.

Hay patrones de uso que varían a lo largo del mes, lo que sugiere oportunidades para optimizar recursos. Viernes en Agosto.

### Recomendaciones

Aplicar políticas de ahorro energético y de servicios en días con ocupación baja.

Considerar el cierre parcial de ciertas áreas o pisos durante esos días.

Monitorear regularmente los patrones de acceso para validar las decisiones tomadas.

Buscar integrar otras fuentes de datos (como consumo energético) para un análisis más profundo en fases futuras.
