# **Proyecto EDA**
### **Análisis exploratorio de patrones de ocupación del edificio para optimización de instalaciones**

### **Objetivo del Proyecto**
Identificar días y horarios con menor ocupación del edificio mediante datos de acceso, con el fin de proponer mejoras en la gestión energética y operativa, optimizando costos relacionados con climatización, iluminación y seguridad.

### **Hipótesis**

Existen días específicos (viernes, lunes, fines de semana, dias antes de dias no laborables o periodos vacacionales) con ocupación mínima, en los cuales mantener la operación regular completa del edificio no se considera rentable.

Mediante la identificación de estos patrones de ocupación es posible calcular una cantidad mínimade recursos necesario para justificar la operación del edificio.

Aplicando ajustes operativos en días de baja ocupación (reducción parcial o cierre total), se lograría un ahorro significativo en costos energéticos.

### **Storytelling**

Ante la creciente adopción del teletrabajo despues de la pandemia, diversos departamentos han manifestado preocupación por la baja ocupación observada en un edificio corporativo, generando costos innecesarios relacionados con energía, climatización, seguridad y limpieza.

En conversaciones recientes, con distintos departamentos, el departamento de Facilities señaló específicamente la necesidad de entender mejor estos patrones para tomar decisiones fundamentadas sobre optimización de recursos. 

Para ello, acordamos colaborar directamente con el área de Seguridad, encargada de administrar el sistema de control de accesos del edificio, desde donde extraeremos los datos necesarios para nuestro análisis.

Este análisis exploratorio (EDA) nos permitirá identificar claramente los patrones de ocupación, generando decisiones con datos.

### **Proximos Pasos**:

Ahora mismo, solo nos dan acceso a estos datos manteniendo el anonimato de las personas que opera en el edificio por la ley de proteccion de datos, por lo que no utilizaremos nombres personales y asignaremos un codigo unico a los datos de las personas. 

La idea de iniciar solo con estos datos es desarrollarlo luego con otro tipo de datos como EVENTOS (Entiendo el impacto de los eventos en los % de ocupacion del edificio y de recursos destinados), Vehiculos (Entendiendo la demanda de cargadores electricos y consumo), y distribucion del personal por las plantas. 


### **Conclusiones y Recomendaciones**:

Identificación de patrones claros y recomendaciones para optimización operativa y energética.
Estimación económica del impacto de las propuestas.


## Dossier - Análisis Exploratorio de Datos (EDA)

### Contexto

La reciente transición hacia modalidades híbridas de trabajo ha resultado en una baja ocupación del edificio corporativo de 12.672 m², distribuido en 6 pisos, un hall principal y 2 sótanos de estacionamiento. Departamentos como Facilities y Seguridad han destacado la importancia de analizar patrones de uso para optimizar recursos operativos como climatización, seguridad e iluminación. Debido a restricciones de acceso, no se dispone de información horaria detallada, solo se cuenta con registros diarios de accesos.

## Análisis Exploratorio

### Análisis Univariante

Distribución diaria de accesos: Identificación de los días con menor ocupación semanal y mensual.

Definiremos el lunes como el primer dia de la semana y el Domingo como el 7mo. 

Frecuencia por piso (si aplica): Análisis del uso relativo de diferentes pisos.

Métricas clave: promedio  , mediana, desviación estándar de accesos diarios.

### Análisis Bivariante

Día de la semana vs. cantidad de accesos: Correlación entre tipo de día (laboral o no) y nivel de ocupación.

Mes vs. cantidad de accesos: Análisis de estacionalidad o variaciones mensuales.

Piso vs. accesos: Comparación de uso entre diferentes niveles del edificio.

### Análisis Multivariante

Día-mes-piso (si aplica): Combinación de variables para entender patrones más complejos.

Clusterización de días según cantidad de accesos y distribución por piso: Segmentación de días similares.

Análisis predictivo simple: Estimación de días con baja ocupación futura en base a tendencias.

### Visualizaciones Relevantes

Heatmaps (día vs. semana): Identificación visual de los días más y menos concurridos.

Histogramas: Distribución de accesos diarios.

Gráficos de líneas: Tendencias de ocupación a lo largo del tiempo.

Gráficos de barras: Comparación de accesos por piso o por mes.

Gráficos de clusters: Visualización de días agrupados por patrones similares.

### Conclusiones

Se identifican ciertos días de la semana (como viernes o lunes) con una reducción constante en el número de accesos.

Hay patrones de uso que varían a lo largo del mes, lo que sugiere oportunidades para optimizar recursos.

Algunos pisos presentan menor uso constante, lo que permite considerar ajustes operativos por nivel.

### Recomendaciones

Aplicar políticas de ahorro energético y de servicios en días con ocupación baja.

Considerar el cierre parcial de ciertas áreas o pisos durante esos días.

Monitorear regularmente los patrones de acceso para validar las decisiones tomadas.

Buscar integrar otras fuentes de datos (como consumo energético) para un análisis más profundo en fases futuras.
